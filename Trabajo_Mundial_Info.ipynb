```python
#Se genera Array con la info a filtar, agrupar y graficar
Info_general = info.loc[:,['Region', 'Pais', 'Ano' ,'Poblacion total','PIB', 'Uso de energia']]

#Se liminan valores en nulo por que afectan los totales
Info_general = Info_general[ ( Info_general['PIB'].notnull() ) & (Info_general['Uso de energia'].notnull() ) ]  

#Se elimina fecha y se deja solo el campo de año 
Info_general['Ano'] = Info_general['Ano'].str.replace('-12-01','')


Info_general_agrupada = Info_general.sort_values(by=['Region','Pais','Ano']).groupby(['Region','Pais','Ano']).sum().reset_index()
```

```python
Info_general_agrupada
```

```python
#Se traen librerias para graficar
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#Se separa un array con los años
anos_consulta = Info_general_agrupada[['Ano']].groupby(['Ano']).mean().reset_index()
anos_consulta
```

```python
#Se calcula promedios por regiones
Consumo_energia_agrupada_promedio = Info_general_agrupada.groupby(["Region","Ano"])["Uso de energia"].transform("mean")
Info_general_agrupada['Consumo energia media por región'] = Consumo_energia_agrupada_promedio

PIB_agrupada_promedio = Info_general_agrupada.groupby(["Region", "Ano"])["PIB"].transform("mean")
Info_general_agrupada['PIB media por región'] = PIB_agrupada_promedio

Info_general_agrupada["Poblacion total"] = Info_general_agrupada["Poblacion total"].astype("int64")
Info_general_agrupada["Consumo energia media por región"] = Info_general_agrupada["Consumo energia media por región"].astype("int64")
Info_general_agrupada["Uso de energia"] = Info_general_agrupada["Uso de energia"].astype("int64")
Info_general_agrupada["PIB"] = Info_general_agrupada["PIB"].astype("int64")
Info_general_agrupada["PIB media por región"] = Info_general_agrupada["PIB media por región"].astype("int64")

     
```

```python
Consumo_energia_agrupada_promedio
```

```python
Info_general_agrupada[ (Info_general_agrupada['Region'] == 'Africa')&( Info_general_agrupada['Ano'] == '2000' ) ]
```

```python
Info_general_grafica = Info_general_agrupada.loc[:,["Region","Ano","Uso de energia","Consumo energia media por región","PIB", "PIB media por región"]]
Info_general_grafica = Info_general_grafica.groupby(["Region","Ano","Consumo energia media por región","PIB media por región"]).sum().reset_index()
```

```python
Info_general_grafica
```

```python
 #Se grafica el consumo total de energia por año en las regiones
len(anos_consulta)
pos_for = 0
data_grafica = Info_general_grafica[ (Info_general_grafica['Ano'] == anos_consulta['Ano'][pos_for]) ]

plt.title("Consumo Por Region año "+anos_consulta['Ano'][pos_for])

x_data_axis = data_grafica["Region"]
y_data_axis = data_grafica["Uso de energia"]
z_data_axis = data_grafica["Consumo energia media por región"]

plt.bar(np.arange(len(x_data_axis)) - 0.2, y_data_axis, 0.4, label = "Consumo de energia Total")
plt.bar(np.arange(len(x_data_axis)) + 0.2, z_data_axis, 0.4, label = "consumo de energia medio")
plt.xticks(np.arange(len(x_data_axis)), x_data_axis)
plt.xlabel("Región")
plt.ylabel("Consumo")
plt.legend()
plt.show()

  

  
   
```
